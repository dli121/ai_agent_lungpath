```python
# 🧠 Healthcare AI Agent - Starter (ReAct Framework)
# This notebook demonstrates a basic reasoning → action → observation loop
# Participants are encouraged to extend this with LLMs, APIs, and memory

import pandas as pd
import os

# -------------------------------
# 🔍 Check Dataset Path (DEBUG)
# -------------------------------
print("📂 Available files:\n")

for root, dirs, files in os.walk('/kaggle/input'):
    print(root)
    for file in files:
        print("   ", file)

# -------------------------------
# 📦 Load Dataset (FIXED PATH)
# -------------------------------

DATA_PATH = '/kaggle/input/competitions/healthcare-ai-agent-hackathon-build-real-world-agents/healthcare_ai_agent_dataset.csv'

df = pd.read_csv(DATA_PATH)

print("\n✅ Dataset Loaded Successfully!\n")
print(df.head())

# -------------------------------
# 🧠 Agent Core Functions
# -------------------------------

def reason(symptoms):
    """Step 1: Reason about symptoms"""
    print("\n[THOUGHT] Analyzing symptoms...")

    if "chest pain" in symptoms.lower():
        return "Possible serious cardiovascular issue. Need urgent clarification."
    elif "fever" in symptoms.lower():
        return "Likely infection-related. Ask follow-up symptoms."
    elif "breathing" in symptoms.lower():
        return "Respiratory concern. Check severity."
    else:
        return "Insufficient data. Ask more questions."

# -------------------------------

def act(symptoms):
    """Step 2: Take action (search dataset)"""
    print("[ACTION] Searching dataset for matching cases...")

    results = df[df['symptoms'].str.contains(symptoms, case=False, na=False)]

    if len(results) > 0:
        return results.iloc[0]
    else:
        return None

# -------------------------------

def observe(result):
    """Step 3: Observe results"""
    print("[OBSERVATION] Processing retrieved data...")

    if result is not None:
        return {
            "follow_up": result['follow_up_question'],
            "conditions": result['possible_conditions'],
            "action": result['recommended_action']
        }
    else:
        return None

# -------------------------------
# 🔁 ReAct Agent Loop
# -------------------------------

def healthcare_agent(user_input):
    print("\n" + "="*50)
    print(f"👤 User Input: {user_input}")

    # Step 1: Reason
    thought = reason(user_input)
    print(f"[THOUGHT] {thought}")

    # Step 2: Act
    result = act(user_input)

    # Step 3: Observe
    observation = observe(result)

    # Final Output
    print("\n🤖 Agent Response:")

    if observation:
        print(f"➡️ Follow-up Question: {observation['follow_up']}")
        print(f"➡️ Possible Conditions: {observation['conditions']}")
        print(f"➡️ Recommended Action: {observation['action']}")
    else:
        print("⚠️ Not enough data. Please provide more detailed symptoms.")

# -------------------------------
# 🧪 Test Cases
# -------------------------------

healthcare_agent("fever")
healthcare_agent("chest pain")
healthcare_agent("breathing difficulty")
healthcare_agent("back pain")
```
